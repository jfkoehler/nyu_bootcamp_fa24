# Using interactive widgets with the notebook

The `ipywidgets` library contains a variety of tools for making interactive elements in the notebook environment. Some basic examples are shown below, and your goal is to incorporate widgets into a plot to add interactivity. Be sure to check out the documentation [here](https://ipywidgets.readthedocs.io/en/stable/) for further examples and inspiration.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import ipywidgets as widgets
from ipywidgets import interact, interactive
````

### Using a Slider

````python
def f(x):
    return x
````

````python
interact(f, x = widgets.FloatSlider(min = -3, max = 3, step = 0.1));
````

````python
interact(f, x=['apples','oranges']);
````

### Alternative Approaches

````python
#using a decorator
@interact(x = 10)
def f(x):
    return x**2
````

````python
#interactive returns object to then be displayed
from IPython.display import display
def f(a, b):
    display(a + b)
    return a+b
````

````python
w = interactive(f, a=10, b=20)
````

````python
display(w)
````

````python
#controlling layout with HBox
a = widgets.IntSlider()
b = widgets.IntSlider()
c = widgets.IntSlider()
ui = widgets.HBox([a, b, c])
def f(a, b, c):
    print((a, b, c))

out = widgets.interactive_output(f, {'a': a, 'b': b, 'c': c})

display(ui, out)
````

````python
def f(m, b):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-5, 5)
    plt.grid()
    plt.axhline(color = 'black')
    plt.axvline(color = 'black')
    plt.title(f'$f(x) = {m}x + {b}$')
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
````

### Working with Gapminder Data

````python
df = pd.read_csv('https://raw.githubusercontent.com/jfkoehler/nyu_bootcamp_fa24/refs/heads/main/data/gapminder_all.csv')
````

````python
df.head()
````

````python
continent_color = {'Africa': 'turquoise',
                  'Asia': 'red',
                  'Oceania': 'purple',
                  'Americas': 'green',
                  
                  'Europe':'yellow'}
````

````python
df['colors'] = df['continent'].map(continent_color)
````

````python
df['colors'].value_counts()
````

````python
df['continent'].unique()
````

**GOALS**

````python
plt.scatter(np.log(df['gdpPercap_1952']), np.log(df['lifeExp_1952']), c = df['colors'],
           s = df['pop_1952']/10**5, alpha = 0.3, edgecolor = 'black')
plt.grid();
````

#### Animating the Plot

Below, we use the `gif` library to create an animation of the plot. 

````python
#!pip install -U gif
````

````python
import gif
````

````python
@gif.frame
def gapper(year):
    plt.scatter(np.log(df[f'gdpPercap_{year}']), np.log(df[f'lifeExp_{year}']), c = df['colors'],
           s = df[f'pop_{year}']/10**5, alpha = 0.3, edgecolor = 'black')
    plt.grid();
    plt.xlim(5, 12)
    plt.ylim(3, 5)
    plt.title(f'Health vs. Wealth in {year}')
````

````python
gapper(1952)
````

````python
interact(gapper, year = widgets.IntSlider(min = 1952, max = 2007, step = 5))
````

#### Making a `gif`

````python
frames = []
for i in range(1952, 2012, 5):
    frame = gapper(i)
    frames.append(frame)
````

````python
gif.save(frames, 'gapminder.gif', duration=100, unit="ms", between="frames")
````
